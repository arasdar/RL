# Deep Deterministic Policy Gradients (DDPG)
---
In this notebook, we train DDPG with OpenAI Gym's Cart-pole environment.

### 1. Import the Necessary Packages

```python
import gym
import random
import torch
import numpy as np
from collections import deque
import matplotlib.pyplot as plt
%matplotlib inline

from ddpg_agent import Agent
```

### 2. Instantiate the Environment and Agent

```python
#env = gym.make('CartPole-v0')
env = gym.make('CartPole-v1')
```

```python
env.action_space, env.observation_space, env.reward_range
```

```python
print(env.action_space.n, env.reward_range, env.observation_space.shape[0])
```

```python
env.seed(2)
agent = Agent(state_size=4, action_size=2, random_seed=2)
```

### 3. Train the Agent with DDPG

```python
# agent.actor_local.load_state_dict(torch.load('cartpole-checkpoint_actor.pth'))
# agent.critic_local.load_state_dict(torch.load('cartpole-checkpoint_critic.pth'))
total_reward_deque = deque(maxlen=100)
for ep in range(111111):
    state = env.reset()
    agent.reset()
    total_reward = 0
    while True:
        action = agent.act(state)
        #print(action)
        action_class = np.argmax(action)
        #print(action_class)
        next_state, reward, done, _ = env.step(action_class)
        agent.step(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward
        if done:
            break 
    total_reward_deque.append(total_reward)
    print('\rEpisode {}\tAverage Score: {:.2f}'.format(ep, np.mean(total_reward_deque)))   
    if np.mean(total_reward_deque) >= 500:
        torch.save(agent.actor_local.state_dict(), 'cartpole-checkpoint_actor.pth')
        torch.save(agent.critic_local.state_dict(), 'cartpole-checkpoint_critic.pth')
        break
```

### 4. Watch a Smart Agent!

```python
agent.actor_local.load_state_dict(torch.load('pendulum-checkpoint_actor.pth'))
agent.critic_local.load_state_dict(torch.load('pendulum-checkpoint_critic.pth'))

state = env.reset()
while True:
    env.render()
    action = agent.act(state, add_noise=False)
    state, reward, done, _ = env.step(action)
    if done:
        break 

env.close()
```
